```python
for i, item in df.iterrows():
    if pd.notna(item["prevReferenceNo"]):
        matched_prev = df.loc[
            df["referenceNumber"] == item["prevReferenceNo"],
            ["referenceNumber", "organization_value", "totalCostWithoutVAT"]
        ]

        if not matched_prev.empty:
            print("CURRENT ROW:")
            print(item[["referenceNumber", "prevReferenceNo", "nextRefNo"]])
            print("WILL ZERO PREVIOUS CONTRACT:")
            print(matched_prev)
            print("-----")

    if pd.notna(item["nextRefNo"]):
        print("CURRENT ROW:")
        print(item[["referenceNumber", "prevReferenceNo", "nextRefNo"]])
        print("WILL ALSO ZERO THIS CURRENT CONTRACT BECAUSE IT HAS nextRefNo:")
        print(df.loc[[i], ["referenceNumber", "organization_value", "totalCostWithoutVAT"]])
        print("-----")
```

```python
print(df.loc[df["referenceNumber"].isin(df["prevReferenceNo"].dropna()), ["referenceNumber", "totalCostWithoutVAT"]])
print(df.loc[df["nextRefNo"].notna(), ["referenceNumber", "totalCostWithoutVAT"]])
```
