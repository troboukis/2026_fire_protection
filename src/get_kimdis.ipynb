```python
import pandas as pd
import requests

ref_number = "24SYMV015476714"
url = "https://cerpp.eprocurement.gov.gr/khmdhs-opendata/contract?page=0"

r = requests.post(
    url,
    json={"referenceNumber": ref_number},
    headers={"Accept": "application/json", "Content-Type": "application/json"},
    timeout=30,
)
r.raise_for_status()

data = r.json()
df_ref = pd.json_normalize(data.get("content", []))
df_ref
```
