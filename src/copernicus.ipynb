```python
import requests
import pandas as pd

url = "https://api.effis.emergency.copernicus.eu/rest/2/burntareas/current/"
params = {
    "country": "EL",
    "firedate__gte": "2025-01-01T00:00:00",
    "firedate__lte": "2025-12-31T23:59:59",
    "ordering": "-lastupdate,-area_ha",
    "limit": 100,
}

headers = {
    "Accept": "application/json",
    "Origin": "https://forest-fire.emergency.copernicus.eu",
    "Referer": "https://forest-fire.emergency.copernicus.eu/",
    "User-Agent": "Mozilla/5.0"
}

rows = []
next_url = url

while next_url:
    r = requests.get(next_url, params=params if next_url == url else None, headers=headers, timeout=30)
    r.raise_for_status()
    payload = r.json()

    rows.extend(payload["results"])
    next_url = payload["next"]

df = pd.DataFrame(rows)
```

```python
df.to_csv("copernicus_2025.csv", index=False)
```

```python
df['firedate'] = pd.to_datetime(df['firedate'], format='ISO8601', utc=True)
```
